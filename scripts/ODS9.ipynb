```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

import plotly.io as pio
```

# Indicadores Globales

## Indicador 9.3.2:
### Proporción de las pequeñas industrias que han obtenido un préstamo o una línea de crédito

```python
# Path

path = "../datos"
# drive_path = "/content/drive/MyDrive/HackODS/data"

# Importamos
df_932_numero_unidades = pd.read_csv(f"{path}/9.3.2_dc_1920_es.csv", usecols=[0,1])
df_932_total_unidades = pd.read_csv(f"{path}/9.3.2_dc_1921_es.csv", usecols=[0,1])
df_932_proporcion_unidades = pd.read_csv(f"{path}/9.3.2_sh_es.csv", usecols=[0,1])

# Renombramos
df_932_numero_unidades.rename(columns={df_932_numero_unidades.columns[1]: "Numero de unidades"}, inplace=True)
df_932_total_unidades.rename(columns={df_932_total_unidades.columns[1]: "Total de unidades"}, inplace=True)
df_932_proporcion_unidades.rename(columns={df_932_proporcion_unidades.columns[1]: "Proporcion de unidades"}, inplace=True)

# Unimos y ordenamos
df_932 = df_932_numero_unidades.merge(df_932_total_unidades, on="Periodo", how="inner")
df_932 = df_932.merge(df_932_proporcion_unidades, on="Periodo", how="inner")
df_932 = df_932.sort_values(by='Periodo').reset_index(drop=True)
```

```python

fig = px.line(
    df_932, 
    x='Periodo', 
    y='Proporcion de unidades',
    title='Caída del Indicador 9.3.2: Pequeñas Industrias con Crédito',
    markers=True,
    text='Proporcion de unidades'
)

fig.update_traces(
    line_color='#d62728', 
    line_width=3,
    marker=dict(size=10),
    texttemplate='%{text:.2f}%',
    textposition='top center',
    textfont=dict(family="Arial", size=12, color="black")
)

fig.update_layout(
    xaxis_title='Año',
    yaxis_title='Proporción',
    xaxis=dict(
        tickmode='linear', 
        tick0=df_932['Periodo'].min(), 
        dtick=1
    ),
    yaxis=dict(showgrid=True, gridcolor='rgba(0,0,0,0.1)'),
    plot_bgcolor='white',
    title_x=0.5,
    height=600
)

fig.show()
```

```python
numero_unidades = pd.to_numeric(df_932['Numero de unidades'].astype(str).str.replace(',', '', regex=False))
total_unidades = pd.to_numeric(df_932['Total de unidades'].astype(str).str.replace(',', '', regex=False))
proporcion_unidades = pd.to_numeric(df_932['Proporcion de unidades'])
periodos = df_932['Periodo'].astype(str)

fig = make_subplots(specs=[[{"secondary_y": True}]])

fig.add_trace(
    go.Bar(
        x=periodos,
        y=total_unidades,
        name='Total de unidades',
        marker_color='#1f77b4',
        text=[f"{v:,.0f}" for v in total_unidades],
        textposition='outside',
        insidetextanchor='end',
        textfont=dict(size=13, color='#1f77b4', weight='bold'),
        offsetgroup=1
    ),
    secondary_y=False
)

fig.add_trace(
    go.Bar(
        x=periodos,
        y=numero_unidades,
        name='Unidades con préstamo',
        marker_color="#eca601",
        text=[f"{v:,.0f}" for v in numero_unidades],
        textposition='outside',
        cliponaxis=False,
        textfont=dict(size=15, color='#eca601', weight='bold'),
        offsetgroup=1
    ),
    secondary_y=False
)

fig.add_trace(
    go.Scatter(
        x=periodos,
        y=proporcion_unidades,
        name='Proporción',
        mode='lines+markers+text',
        marker=dict(symbol='diamond', size=12, color='#d62728'),
        line=dict(color='#d62728', width=4),
        text=[f"{v:.2f}%" for v in proporcion_unidades],
        textposition='top center',
        cliponaxis=False,
        textfont=dict(color='#282828', size=13, weight='bold')
    ),
    secondary_y=True
)

fig.update_layout(
    title=dict(
        text='<b>Brecha creciente: Crecimiento Industrial vs. Acceso al Crédito</b>',
        x=0.5,
        font=dict(size=18)
    ),
    xaxis=dict(title='Año', tickmode='linear'),
    yaxis=dict(title='Número de Unidades', showgrid=True, gridcolor='rgba(0,0,0,0.05)'),
    yaxis2=dict(title='Proporción (%)', range=[0, 35], color='#d62728', overlaying='y', side='right'),
    barmode='overlay',
    plot_bgcolor='white',
    margin=dict(t=120, b=50, l=50, r=50),
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="center", x=0.5),
    uniformtext=dict(minsize=12, mode='show'),
    height=700
)

fig.show()
```

### Notas:
Lo que nos dicen estas gráficas es que han aumentado las industrias pero ha disminuido el porcentaje a las cuales se les otorga un préstamo para seguir creciendo

## Indicador 9.5.1:

### Gasto federal en investigación científica y desarrollo experimental como proporción del PIB

```python
# Importamos
df_951_gasto_investigacion = pd.read_csv(f"{path}/9.5.1_dc_501_es.csv", usecols=[0,1])
df_951_pbi_anual = pd.read_csv(f"{path}/9.5.1_dc_502_es.csv", usecols=[0,1])
df_gasto_proporcional = pd.read_csv(f"{path}/9.5.1_sh_es.csv", usecols=[0,1])

# Renombramos

df_951_gasto_investigacion.rename(columns={df_951_gasto_investigacion.columns[1]: "Gasto en investigacion"}, inplace=True)
df_951_pbi_anual.rename(columns={df_951_pbi_anual.columns[1]: "PBI anual"}, inplace=True)
df_gasto_proporcional.rename(columns={df_gasto_proporcional.columns[1]: "Gasto proporcional"}, inplace=True)

# Unimos y ordenamos
df_951 = df_951_gasto_investigacion.merge(df_951_pbi_anual, on="Periodo", how="inner")
df_951 = df_951.merge(df_gasto_proporcional, on="Periodo", how="inner")
df_951 = df_951.sort_values(by='Periodo').reset_index(drop=True)
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import pandas as pd

# Limpieza de datos
df_951['Gasto en investigacion'] = pd.to_numeric(
    df_951['Gasto en investigacion'].astype(str).str.replace(',', '', regex=False)
)
df_951['PBI anual'] = pd.to_numeric(
    df_951['PBI anual'].astype(str).str.replace(',', '', regex=False)
)
df_951['Gasto proporcional'] = pd.to_numeric(df_951['Gasto proporcional'])
periodos = df_951['Periodo'].astype(str)

fig = make_subplots(specs=[[{"secondary_y": True}]])

# Barra ANCHA (Fondo - PBI anual)
fig.add_trace(
    go.Bar(
        x=periodos,
        y=df_951['PBI anual'],
        name='PBI anual (MDP)',
        marker_color='#1f77b4',
        text=[f"{v:,.0f}" for v in df_951['PBI anual']],
        textposition='inside',
        insidetextanchor='end',
        textfont=dict(size=11, color='white', weight='bold'),
        offsetgroup=1
    ),
    secondary_y=False
)

# Barra DELGADA (Frente - Gasto en investigación)
fig.add_trace(
    go.Bar(
        x=periodos,
        y=df_951['Gasto en investigacion'],
        name='Gasto en investigación (MDP)',
        marker_color='#eca601',
        opacity=0.9,
        width=0.3,
        text=[f"{v:,.0f}" for v in df_951['Gasto en investigacion']],
        textposition='outside',
        cliponaxis=False,
        textfont=dict(size=11, color='#eca601', weight='bold'),
        offsetgroup=1
    ),
    secondary_y=False
)

# Línea - Gasto proporcional (eje secundario)
fig.add_trace(
    go.Scatter(
        x=periodos,
        y=df_951['Gasto proporcional'],
        name='Gasto proporcional (% PBI)',
        mode='lines+markers+text',
        marker=dict(symbol='diamond', size=12, color='#d62728'),
        line=dict(color='#d62728', width=4),
        text=[f"{v:.3f}%" for v in df_951['Gasto proporcional']],
        textposition='top center',
        cliponaxis=False,
        textfont=dict(color='#282828', size=11, weight='bold')
    ),
    secondary_y=True
)

fig.update_layout(
    title=dict(
        text='<b>Evolución del Gasto Federal en I+D como % del PBI</b>',
        x=0.5,
        font=dict(size=18)
    ),
    xaxis=dict(title='Año', tickmode='linear'),
    yaxis=dict(title='Millones de pesos (MDP)', showgrid=True, gridcolor='rgba(0,0,0,0.05)'),
    yaxis2=dict(title='% del PBI', color='#d62728', overlaying='y', side='right'),
    barmode='overlay',
    plot_bgcolor='white',
    margin=dict(t=120, b=50, l=50, r=50),
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="center", x=0.5),
    uniformtext=dict(minsize=10, mode='show'),
    height=700
)

fig.show()
```

## Indicador 9n.2.1:
### Multiplicador del subsidio a la investigación científica y desarrollo experimental financiada por las empresas

```python
# Importamos
df_9n21_valor_t_proyectos_autorizados = pd.read_csv(f"{path}/9n.2.1_dc_2112_es.csv", usecols=[0,1])
df_9n21_valor_t_estimulo_fiscal = pd.read_csv(f"{path}/9n.2.1_dc_2113_es.csv", usecols=[0,1])
df_9n21_multiplicador_subsidio = pd.read_csv(f"{path}/9n.2.1_sh_es.csv", usecols=[0,1])

# Renombramos

df_9n21_valor_t_proyectos_autorizados.rename(columns={df_9n21_valor_t_proyectos_autorizados.columns[1]: "Valor total proyectos autorizados"}, inplace=True)
df_9n21_valor_t_estimulo_fiscal.rename(columns={df_9n21_valor_t_estimulo_fiscal.columns[1]: "Valor total estímulo fiscal"}, inplace=True)
df_9n21_multiplicador_subsidio.rename(columns={df_9n21_multiplicador_subsidio.columns[1]: "Multiplicador del subsidio"}, inplace=True)

# Unimos y ordenamos
df_9n21 = df_9n21_valor_t_proyectos_autorizados.merge(df_9n21_valor_t_estimulo_fiscal, on="Periodo", how="inner")
df_9n21 = df_9n21.merge(df_9n21_multiplicador_subsidio, on="Periodo", how="inner")
df_9n21 = df_9n21.sort_values(by='Periodo').reset_index(drop=True)
```

```python
autorizado  = pd.to_numeric(df_9n21['Valor total proyectos autorizados'].astype(str).str.replace(',', '', regex=False))
estimulo    = pd.to_numeric(df_9n21['Valor total estímulo fiscal'].astype(str).str.replace(',', '', regex=False))
multiplicador = pd.to_numeric(df_9n21['Multiplicador del subsidio'])
periodos    = df_9n21['Periodo'].astype(str)

fig = make_subplots(specs=[[{"secondary_y": True}]])

# Barra ANCHA - Valor autorizado (fondo)
fig.add_trace(
    go.Bar(
        x=periodos,
        y=autorizado,
        name='Valor total proyectos autorizados',
        marker_color='#aec6e8',
        text=[f"${v:,.1f}M" for v in autorizado],
        textposition='inside',
        insidetextanchor='end',
        textfont=dict(size=11, color='#2c5f8a'),
        offsetgroup=1
    ),
    secondary_y=False
)

# Barra DELGADA - Estímulo fiscal (frente)
fig.add_trace(
    go.Bar(
        x=periodos,
        y=estimulo,
        name='Valor total estímulo fiscal',
        marker_color='#1f77b4',
        width=0.35,
        text=[f"${v:,.1f}M" for v in estimulo],
        textposition='outside',
        cliponaxis=False,
        textfont=dict(size=12, color='#1f77b4', weight='bold'),
        offsetgroup=1
    ),
    secondary_y=False
)

# Línea - Multiplicador (eje secundario)
fig.add_trace(
    go.Scatter(
        x=periodos,
        y=multiplicador,
        name='Multiplicador del subsidio',
        mode='lines+markers+text',
        line=dict(color='#d62728', width=3.5),
        marker=dict(symbol='diamond', size=11, color='#d62728'),
        text=[f"{v:.2f}x" for v in multiplicador],
        textposition='top center',
        cliponaxis=False,
        textfont=dict(color='#282828', size=12, weight='bold')
    ),
    secondary_y=True
)

# Línea de tendencia del multiplicador
z = np.polyfit(range(len(df_9n21)), multiplicador, 1)
tendencia = np.poly1d(z)(range(len(df_9n21)))

fig.add_trace(
    go.Scatter(
        x=periodos,
        y=tendencia,
        name='Tendencia multiplicador',
        mode='lines',
        line=dict(color='#d62728', width=1.5, dash='dash'),
        opacity=0.5
    ),
    secondary_y=True
)

fig.update_layout(
    title=dict(
        text='<b>Estímulo Fiscal: Mayor gasto, menor eficiencia del multiplicador</b>',
        x=0.5,
        font=dict(size=17)
    ),
    xaxis=dict(title='Año', tickmode='linear'),
    yaxis=dict(
        title='Millones de pesos',
        showgrid=True,
        gridcolor='rgba(0,0,0,0.05)'
    ),
    yaxis2=dict(
        title='Multiplicador (x veces)',
        range=[3, 7],
        color='#d62728',
        overlaying='y',
        side='right',
        showgrid=False
    ),
    barmode='overlay',
    plot_bgcolor='white',
    margin=dict(t=120, b=60, l=60, r=60),
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="center", x=0.5),
    uniformtext=dict(minsize=11, mode='show'),
    height=680
)

fig.show()
```
