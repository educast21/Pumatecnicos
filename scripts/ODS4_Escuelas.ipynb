```python
import plotly.graph_objects as go
import pandas as pd
import plotly.express as px
import json
import requests
from plotly.subplots import make_subplots
```

```python
# Electricidad
path = "../datos"
df = pd.read_excel(path + "/4.A.1.A_sh_es.xls", sheet_name=0, header=None)
df = df.iloc[7:-6]
df.columns = df.iloc[0]
df = df[2:]
df_elect = df

# Internet
df = pd.read_excel(path + "/4.A.1.B_sh_es.xls", sheet_name=0, header=None)
df = df.iloc[7:-6]
df.columns = df.iloc[0]
df = df[2:]
df_internet = df

# Equipo de computo
df = pd.read_excel(path + "/4.A.1.C_sh_es.xls", sheet_name=0, header=None)
df = df.iloc[7:-6]
df.columns = df.iloc[0]
df = df[2:]
df_comput = df

# Agua potable
df = pd.read_excel(path + "/4.A.1.E_sh_es.xls", sheet_name=0, header=None)
df = df.iloc[7:-6]
df.columns = df.iloc[0]
df = df[2:]
df_agua = df

# Sanitarios
df = pd.read_excel(path + "/4.A.1.F_sh_es.xls", sheet_name=0, header=None)
df = df.iloc[7:-6]
df.columns = df.iloc[0]
df = df[2:]
df_sanit = df

# Lavabos
df = pd.read_excel(path + "/4.A.1.G_sh_es.xls", sheet_name=0, header=None)
df = df.iloc[7:-6]
df.columns = df.iloc[0]
df = df[2:]
df_lavab = df

# columnas de excel: C, D, F, G, H, J, K, L, N, O, P, R, S, T, V, W, X, Z, AA, AB, AD
indices = [1,2, 3, 5, 6, 7, 9, 10, 11, 13, 14, 15, 17, 18, 19, 21, 22, 23, 25, 26, 27, 29]

# Electricidad
df_elect_nal = df_elect.iloc[:, indices]
df_elect_nal = df_elect_nal.iloc[:-32,1:]
df_elect_nal = df_elect_nal.values.flatten().reshape(7,3)
anios = ['2025','2024' ,'2023','2022', '2021','2020','2019']
df_elect_nal = pd.DataFrame(df_elect_nal.T, columns=anios)

df_plot_elect = df_elect_nal.melt(
    value_vars=anios,
    var_name='Año',
    value_name='Electricidad'
)

# Internet
df_internet_nal = df_internet.iloc[:, indices]
df_internet_nal = df_internet_nal.iloc[:-32,1:]
df_internet_nal = df_internet_nal.values.flatten().reshape(7,3)
anios = ['2025','2024' ,'2023','2022', '2021','2020','2019']
df_internet_nal = pd.DataFrame(df_internet_nal.T, columns=anios)

df_plot_internet = df_internet_nal.melt(
    value_vars=anios,
    var_name='Año',
    value_name='Internet'
)

# Equipo de computo
df_comput_nal = df_comput.iloc[:, indices]
df_comput_nal = df_comput_nal.iloc[:-32,1:]
df_comput_nal = df_comput_nal.values.flatten().reshape(7,3)
anios = ['2025','2024' ,'2023','2022', '2021','2020','2019']
df_comput_nal = pd.DataFrame(df_comput_nal.T, columns=anios)

df_plot_comput = df_comput_nal.melt(
    value_vars=anios,
    var_name='Año',
    value_name='Cómputo'
)

# Agua potable
df_agua_nal = df_agua.iloc[:, indices]
df_agua_nal = df_agua_nal.iloc[:-32,1:]
df_agua_nal = df_agua_nal.values.flatten().reshape(7,3)
anios = ['2025','2024' ,'2023','2022', '2021','2020','2019']
df_agua_nal = pd.DataFrame(df_agua_nal.T, columns=anios)

df_plot_agua = df_agua_nal.melt(
    value_vars=anios,
    var_name='Año',
    value_name='Agua'
)

# Sanitarios
df_sanit_nal = df_sanit.iloc[:, indices]
df_sanit_nal = df_sanit_nal.iloc[:-32,1:]
df_sanit_nal = df_sanit_nal.values.flatten().reshape(7,3)
anios = ['2025','2024' ,'2023','2022', '2021','2020','2019']
df_sanit_nal = pd.DataFrame(df_sanit_nal.T, columns=anios)

df_plot_sanit = df_sanit_nal.melt(
    value_vars=anios,
    var_name='Año',
    value_name='Sanitarios'
)


# Lavabos
df_lavab_nal = df_lavab.iloc[:, indices]
df_lavab_nal = df_lavab_nal.iloc[:-32,1:]
df_lavab_nal = df_lavab_nal.values.flatten().reshape(7,3)
anios = ['2025','2024' ,'2023','2022', '2021','2020','2019']
df_lavab_nal = pd.DataFrame(df_lavab_nal.T, columns=anios)

df_plot_lavab = df_lavab_nal.melt(
    value_vars=anios,
    var_name='Año',
    value_name='Lavabos'
)

df_plot_elect['Electricidad'] = pd.to_numeric(df_plot_elect["Electricidad"], errors="coerce")
df_plot_elect['Internet'] = pd.to_numeric(df_plot_internet["Internet"], errors="coerce")
df_plot_elect['Cómputo'] = pd.to_numeric(df_plot_comput["Cómputo"], errors="coerce")
df_plot_elect['Agua'] = pd.to_numeric(df_plot_agua["Agua"], errors="coerce")
df_plot_elect['Sanitarios'] = pd.to_numeric(df_plot_sanit["Sanitarios"], errors="coerce")
df_plot_elect['Lavabos'] = pd.to_numeric(df_plot_lavab["Lavabos"], errors="coerce")
df_plot_elect['Nivel'] = ['Primaria', 'Secundaria', 'Preparatoria'] * 7

df = df_plot_elect
df = df.sort_values(by='Año')
```

## Grafico Spider para ver los servicios de cada nivel, de cada estado, en los años 2019 y 2025

```python
# columnas de excel: C, D, F, G, H, J, K, L, N, O, P, R, S, T, V, W, X, Z, AA, AB, AD
indices = [1,2, 3,5, 26, 27, 29]

# Electricidad
df_elect_nal = df_elect.iloc[:, indices]
df_elect_nal = df_elect_nal.iloc[1:,:]

# Internet
df_internet_nal = df_internet.iloc[:, indices]
df_internet_nal = df_internet_nal.iloc[1:,:]

# Equipo de computo
df_comput_nal = df_comput.iloc[:, indices]
df_comput_nal = df_comput_nal.iloc[1:,:]

# Agua potable
df_agua_nal = df_agua.iloc[:, indices]
df_agua_nal = df_agua_nal.iloc[1:,:]

# Sanitarios
df_sanit_nal = df_sanit.iloc[:, indices]
df_sanit_nal = df_sanit_nal.iloc[1:,:]

# Lavabos
df_lavab_nal = df_lavab.iloc[:, indices]
df_lavab_nal = df_lavab_nal.iloc[1:,:]

columnas = ['Entidad',
    '2025_Primaria', '2025_Secundaria', '2025_Preparatoria',
    '2019_Primaria', '2019_Secundaria', '2019_Preparatoria']

df_elect_nal.columns = columnas
df_internet_nal.columns = columnas
df_comput_nal.columns = columnas
df_agua_nal.columns = columnas
df_sanit_nal.columns = columnas
df_lavab_nal.columns = columnas

# 2. Aplicamos MELT
df_long = df_elect_nal.melt(
    id_vars=['Entidad'],
    var_name='Año_Nivel',
    value_name='Porcentaje'
)

# 3. Separamos la columna temporal '2019_Primaria' en dos: 'Año' y 'Nivel'
df_long[['Año', 'Nivel']] = df_long['Año_Nivel'].str.split('_', expand=True)

# 4. Limpiamos y agregamos la columna de Servicio
df_long = df_long.drop(columns=['Año_Nivel'])
df_long['Servicio'] = 'Electricidad'

df_long
datas = [df_internet_nal,df_comput_nal,df_agua_nal,df_sanit_nal,df_lavab_nal]
i=0
for servicio in datas:
  servicio.columns = columnas
  servs = ['Internet', 'Cómputo', 'Agua', 'Sanitarios', 'Lavabos']
  df_aux = servicio.melt(id_vars=['Entidad'],
    var_name='Año_Nivel',
    value_name='Porcentaje')
  df_aux[['Año', 'Nivel']] = df_aux['Año_Nivel'].str.split('_', expand=True)
  df_aux = df_aux.drop(columns=['Año_Nivel'])
  df_aux['Servicio'] = servs[i]
  df_long = pd.concat([df_long,df_aux])
  i = i+1

df = df_long
df['Porcentaje'] = pd.to_numeric(df['Porcentaje'], errors='coerce')

fig = make_subplots(
    rows=1, cols=3,
    specs=[[{'type': 'polar'}, {'type': 'polar'}, {'type': 'polar'}]],
    subplot_titles=("<b>Primaria<b>", "<b>Secundaria<b>", "<b>Preparatoria<b>")
)

estados = df['Entidad'].unique()
niveles = ['Primaria', 'Secundaria', 'Preparatoria']
años = ['2019', '2025']
colores = {'2019': 'rgba(142, 154, 175, 0.1)', '2025': 'rgba(197, 25, 45, 0.1)'}
lineas = {'2019': 'rgba(142, 154, 175, 1)', '2025': 'rgba(197, 25, 45, 0.8)'}

# 2. Agregar Trazas
# El orden será: Estado1(P-19, P-25, S-19, S-25, Prep-19, Prep-25), Estado2(...), etc.
for estado in estados:
    for col_idx, nivel in enumerate(niveles, 1):
        for año in años:
            df_sub = df[(df['Entidad'] == estado) &
                        (df['Año'] == año) &
                        (df['Nivel'] == nivel)]

            fig.add_trace(
                go.Scatterpolar(
                    r=df_sub['Porcentaje'],
                    theta=df_sub['Servicio'],
                    fill='toself',
                    name=f"{año}",
                    fillcolor=colores[año],
                    line=dict(color=lineas[año]),
                    visible=(estado == estados[0]),
                    showlegend=(col_idx == 1 and estado == estados[0]) # Leyenda solo una vez
                ),
                row=1, col=col_idx
            )

# 3. Lógica del Botón
botones = []
trazas_por_estado = len(niveles) * len(años) # 3 niveles * 2 años = 6 trazas por estado

for i, estado in enumerate(estados):
    visibilidad = [False] * len(fig.data)
    inicio = i * trazas_por_estado
    visibilidad[inicio : inicio + trazas_por_estado] = [True] * trazas_por_estado

    botones.append(dict(
        label=estado,
        method="update",
        args=[{"visible": visibilidad},
              {"title": f"Infraestructura Educativa en {estado} 2019 (en gris) vs. 2025 (en rojo)"}]
    ))

# 4. Diseño Final
fig.update_layout(
    updatemenus=[{
        "buttons": botones,
        "direction": "down",
        "showactive": True,
        "x": 0.95,
        "xanchor": "center",
        "y": 1.2
    }],
    title_text=f"Infraestructura Educativa en {estados[0]} 2019 (en gris) vs. 2025 (en rojo)",
    polar=dict(radialaxis=dict(visible=True, range=[0, 100],tickfont=dict(size=8))),
    polar2=dict(radialaxis=dict(visible=True, range=[0, 100],tickfont=dict(size=8))),
    polar3=dict(radialaxis=dict(visible=True, range=[0, 100],tickfont=dict(size=8))),
    width=1500,
    height=500
)

fig.show()
```
