## 4r.6.1.a(2)  Tasa de alfabetización de las personas de 15 años o más, desglosada por sexo y subpoblaciones [ENIGH]

```python
import plotly.graph_objects as go
import pandas as pd
import plotly.express as px
import json
import requests
```

## Serie de tiempo de porcentaje de alfabetización a nivel nacional

```python
df = pd.read_excel("/content/4R.6.1.A(2)_sh_es.xls", sheet_name=0, header=None)
df = df.iloc[6:-42]
df.columns = df.iloc[0]
df = df[2:]
df = df.iloc[:, [2,3,4, 32,33,34, 62,63,64, 92,93,94, 122,123,124]]
df = df.drop(df.index[0])
df = df.values.flatten().reshape(5, 3)
df = pd.DataFrame(df.T, columns=['2024', '2022', '2020', '2018', '2016'])

df_plot = df.melt(
    value_vars=['2016', '2018', '2020', '2022','2024'],
    var_name='Año',
    value_name='Porcentaje'
)
df_plot['Categoría'] = ['Total', 'Hombre', 'Mujer'] * 5
df_plot['Porcentaje'] = pd.to_numeric(df_plot["Porcentaje"], errors="coerce")


colores = {
    "Total": "#7c0229",
    "Hombre": "#5682f4",
    "Mujer": "#f60e57"
}

# 3. Crear la gráfica de líneas
fig = px.line(
    df_plot,
    x='Año',
    y='Porcentaje',
    color='Categoría',
    color_discrete_map=colores,# Esto crea una serie (línea) por cada categoría
    title='Proporción de la población alfabeta de 15 años o más,<br>'+ 'del año 2016 al 2022, a nivel nacional',
    markers=True # Para que se vean los puntos en cada año
)

# Mejorar el diseño
fig.update_layout(xaxis_type='category') # Para que los años se traten como etiquetas
fig.show()
```

## 4.1.2  Índice de finalización (enseñanza primaria, secundaria y preparatoria) [ENIGH]

```python
df = pd.read_excel("/content/4.1.2_sh_es.xls", sheet_name=0, header=None)
df = df.iloc[7:-39]
df.columns = df.iloc[0]
df = df[2:]
indices = [
    2, 3, 4,       # C, D, E
    8, 9, 10,      # I, J, K
    14, 15, 16,    # O, P, Q
    20, 21, 22,    # U, V, W
    26, 27, 28,    # AA, AB, AC
    32, 33, 34,    # AG, AH, AI
    38, 39, 40,    # AM, AN, AO
    44, 45, 46,    # AS, AT, AU
    50, 51, 52,    # AY, AZ, BA
    56, 57, 58,    # BE, BF, BG
    62, 63, 64,    # BK, BL, BM
    69, 70, 71,    # BR, BS, BT
    74, 75, 76,    # BW, BX, BY
    80, 81, 82,    # CC, CD, CE
    86, 87, 88     # CI, CJ, CK
]
df = df.iloc[:, indices]
df = df.values.flatten().reshape(5,9)
df = pd.DataFrame(df.T, columns=['2024', '2022', '2020', '2018', '2016'])

df_plot = df.melt(
    value_vars=['2016', '2018', '2020', '2022','2024'],
    var_name='Año',
    value_name='Porcentaje'
)

df_plot['Categoría'] = ['Total', 'Hombre', 'Mujer'] * 15
df_plot['TipoEduc'] = (['Primaria']*3 + ['Secundaria']*3 + ['Preparatoria']*3) * 5
df_plot['Porcentaje'] = pd.to_numeric(df_plot["Porcentaje"], errors="coerce")

colores = {
    "Total": "#C5192D",
    "Hombre": "#2b46e5",
    "Mujer": "#ef0796"
}

fig = px.bar(
    df_plot,
    x="TipoEduc",
    y="Porcentaje",
    color="Categoría",
    color_discrete_map= colores,
    barmode="group",
    animation_frame="Año",
    range_y=[0, 110], # Fijamos el eje Y para que no salte al cambiar de año
    title='Proporción de la población que concluyó la educación básica en tiempo,<br>' + 'del año 2016 al 2022, a nivel nacional',
    labels={'TipoEduc': 'Nivel Educativo', 'Porcentaje': 'Porcentaje (%)'},
    category_orders={"Categoría": ["Total", "Hombre", "Mujer"]} # Orden de las barras
)

# 3. Ajustes de diseño
fig.update_layout(
    xaxis_title="<b> Nivel de Escolaridad <b>",
    yaxis_title="<b> Porcentaje (%) <b>",
    legend_title="Categoría"
)

duracion_cuadro = 1500  # 1.5 segundos por año
duracion_transicion = 800 # 0.8 segundos de movimiento

fig.layout.updatemenus[0].buttons[0].args[1]['frame']['duration'] = duracion_cuadro
fig.layout.updatemenus[0].buttons[0].args[1]['transition']['duration'] = duracion_transicion

# Mostrar la gráfica
fig.show()
```

## Serie de tiempo de inscritos en licenciatura

```python
df = pd.read_excel("/content/4R.3.1_sh_es.xls", sheet_name=0, header = None)
df = df.iloc[7:-8] # Eliminar todos los renglones de texto
df.columns = df.iloc[0] # El encabezado es el primer renglón
df = df[1:] # Eliminar el primer renglon (ya es el encazabezado)
df.columns = ["Periodo", "Total", "Hombres", "Mujeres"]
df = df.iloc[::-1].reset_index(drop=True)

colores = {
    "Total": "#C5192D",
    "Hombres": "#2b46e5",
    "Mujeres": "#ef0796"
}

fig = px.line(df, x="Periodo", y=df.columns,
              title='Proporción de la población entre 18 y 22 años <br>'+
              'inscrita en licenciatura, por sexo (Hombre y Mujer) <br>'+
              'del 2015 al 2025',
              markers = True,
              color_discrete_map=colores)
fig.update_xaxes(
    type="category",  # Crucial: Trata las etiquetas como texto, no como tiempo continuo
    showgrid=True,
    gridcolor='LightGray',
    griddash='dot'
)

fig.update_yaxes(
    showgrid=True,
    gridcolor='LightGray',
    griddash='dot'
)
fig.update_layout(
    xaxis_title="Año",
    yaxis_title="Porcentaje",
    legend_title="Categorías",
    plot_bgcolor='white'
)
fig.update_traces(connectgaps=True)
fig.show()
```
