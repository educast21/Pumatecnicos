```python
import plotly.graph_objects as go
import pandas as pd
import plotly.express as px
import json
import requests
from functools import reduce
import numpy as np
path = "../datos"
#path = "/content"
```

```python
df = pd.read_excel(path + "/9.5.2_sh_es.xls", sheet_name=0, header = None)
df = df.iloc[9:-5] # Eliminar todos los renglones de texto, solo dejar el nacional
df.columns = df.iloc[0] # El encabezado es el primer renglón
#df = df[1:] # Eliminar el primer renglon (ya es el encazabezado)
df = df.iloc[:, [1, 3,4]]
df.columns = ["Entidad", "Hombres", "Mujeres"]
df = df.loc[df['Entidad'] == "Estados Unidos Mexicanos"]

data = {
    "Categoría": ["Hombres investigadores", "Mujeres investigadoras"],
    "Investigador@s": [df.values.tolist()[0][1], df.values.tolist()[0][2]]
}

colores = { "Hombres investigadores": "#2b46e5",
           "Mujeres investigadoras": "#ef0796"}

# Crear la figura
fig = px.pie(data, values='Investigador@s', names='Categoría',color='Categoría', color_discrete_map=colores,
             title='En México 2024, por cada millón de habitantes, había en promedio...<br><sup>SECIHTI </sup> ')
fig.update_traces(
    textinfo='value',
    textfont_size=20,     # Tamaño general
    insidetextfont=dict(size=25, color='white') # Tamaño específico para el texto interior
)
# Mostrar la gráfica
fig.show()
```
